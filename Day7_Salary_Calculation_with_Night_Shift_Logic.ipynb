# 深夜残業代の自動計算プログラム

```python
from datetime import datetime
# 基本時給
hourly_wage = 1200
# 深夜手当は25%アップ
night_hourly_wage = 1500
# 19:00 開始
start_time = datetime(2026, 4, 26, 19, 0)
# 23:00 終了
end_time = datetime(2026, 4, 26, 23, 0)

# 深夜残業が始まる時間（22:00）を定義
night_shift_start = datetime(2026, 4, 26, 22, 0)

working_hours = (end_time - start_time).total_seconds() / 3600

if end_time > night_shift_start:
    normal_hours = (night_shift_start - start_time).total_seconds() / 3600
    night_hours = (end_time - night_shift_start).total_seconds() / 3600
else:
    normal_hours = working_hours
    night_hours = 0

normal_pay = normal_hours * hourly_wage
night_pay = night_hours * night_hourly_wage
total_pay = normal_pay + night_pay

print(f"--- 給与明細 ---")
print(f"総労働時間: {working_hours}時間")
print(f"通常労働: {normal_hours}時間 ({int(normal_pay)}円)")
print(f"深夜労働: {night_hours}時間 ({int(night_pay)}円)")
print(f"合計支給額: {int(total_pay)}円")
    
```
