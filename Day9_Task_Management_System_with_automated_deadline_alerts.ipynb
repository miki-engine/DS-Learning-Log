# タスク期限管理システム

```python
from datetime import date, timedelta

# 1. 今日の日付を取得
today = date.today()
# 2. 納品期限の日付
deadline = date(2026, 5, 31)
# 3. タスクリスト [タスク名, 締め切り年, 月, 日]
tasks = [
    ["週次レポート作成", 2026, 5, 2],
    ["クライアント返信", 2026, 4, 29],
    ["プロジェクト準備", 2026, 5, 5],
    ["請求書の確認承認", 2026, 6, 1],
    ["オンボーディング", 2026, 5, 31]
]

print(f"--- 本日の日付 :{today} ---\n")
print("-" * 50)
print(f"{'タスク名':<18}| {'残り日数':} | {'ステータス'}")

for name, y, m, d in tasks:
    deadline = date(y, m, d)
    diff = deadline - today
    days_left = diff.days

    if  days_left < 0:
        icon, status = "❌", "【期限切れ!!】"
    elif days_left <= 1:
        icon, status = "🚨", "【最優先】"
    elif days_left <= 3:
        icon, status = "⚠️", "【緊急】"
    elif days_left <= 7:
        icon, status = "⌛", "【要対応】"
    else:
        icon, status = "✅", "【期限内】"

    print(f"{icon}{name:<12} | {days_left:>3}日残 | {status}")

print("-" * 50)
```
