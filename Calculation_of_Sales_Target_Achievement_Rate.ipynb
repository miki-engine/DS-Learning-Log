# Calculation of Sales Target Achievement Rate

```python
# List of Sales Figures by Store
branch_sales = [85000, 120000, 45000, 100000, 155000]

# Fundraising Goal
TARGET_AMOUNT = 100000

def calculate_achievement(sales):
    """Calculate sales achievement rate against the target."""
    return sales / TARGET_AMOUNT * 100

for sales in branch_sales:
    achievement_rate = calculate_achievement(sales)
    print(achievement_rate)
```
