# 店舗別売上レポートの自動作成

```python
# [作業内容, 金額, 店舗名]
works = [
    ["蛇口修理", 8000, "新宿店"],
    ["エアコン掃除", 12000, "渋谷店"],
    ["屋根補修", 50000, "新宿店"],
    ["鍵交換", 15000, "池袋店"],
    ["草むしり", 5000, "渋谷店"],
    ["家具組み立て", 7000, "新宿店"]
]

sales_report = {}
for name, price, shop in works:
    if shop in sales_report:
        sales_report[shop] += price
    else:
        sales_report[shop] = price

print("店舗別売上集計結果")
for shop, summary in sales_report.items():
    print(f"{shop}: {summary}円")
```
