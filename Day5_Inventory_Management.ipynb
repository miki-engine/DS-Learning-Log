# 在庫管理と自動発注リスト作成

```python
# 1. 商品データ[商品名,現在の在庫数,最低必要数]
inventory_data = [
    ["アジサイの苗", 45, 50],
    ["チューリップの球根", 120, 100],
    ["ナノハナの種", 15, 30],
    ["タンポポの苗", 80, 50],
    ["ヒマワリの種", 5, 20]
]

order_list = [] # 発注が必要な商品を入れる空リスト

print("--- 在庫チェック開始 ---")

# 2. 在庫チェックのループ
for item in inventory_data:
    name = item[0]
    stock = item[1]
    threshold = item[2]

    if stock < threshold:
        shortage = threshold - stock
        print(f"【要発注】{name}: 残り{stock}個({shortage}個不足しています)")
        order_list.append([name, shortage])
    else:
        print(f"【在庫OK】{name}; 残り{stock}個")

# 3. 最終的な発注リストの表示
print("\n--- 本日の発注依頼書 ---")
if not order_list:
    print("発注が必要な商品はありません")
else:
    for order in order_list:
        print(f"商品名: {order[0]:<10} | 発注数: {order[1]:>3}個")
    
```
