# リピーター顧客の自動特定

```python
# [顧客名, 作業日, 支払金額]
customer_logs = [
    ["田中", "2026-04-01", 15000],
    ["佐藤", "2026-04-02", 8000],
    ["田中", "2026-04-10", 12000],
    ["鈴木", "2026-04-15", 5000],
    ["佐藤", "2026-04-20", 20000], #
    ["高橋", "2026-04-25", 10000]
]

customer_counts = {}

for name, dt, amount in customer_logs:
    if name in customer_counts:
        customer_counts[name] += 1
    else:
        customer_counts[name] = 1

print("--- リピーターリスト ---")
for name, count in customer_counts.items():
    if count >= 2:
        print(f"{name}さん: {count}回利用")
```
