# 顧客の購入金額によるランク自動判定

```python
# 1. 顧客データ（名前と今月の購入総額）
customer_data = [
    ["Tanaka", 85000],
    ["Sato", 12000],
    ["Suzuki", 3000],
    ["Takahashi", 150000],
    ["Ito", 45000]
]

print("---　顧客ランク判定システム ---")

# 2. 判定と分類
for customer in customer_data:
    name = customer[0]
    amount = customer[1]

    # 複数条件の分岐（elif)の練習
    if amount >= 100000:
        rank = "VIP顧客"
    elif amount >= 50000:
        rank = "優良顧客"
    elif amount >= 10000:
        rank = "一般顧客"
    else:
        rank = "育成対象"

    print(f"顧客名: {name:10} | 購入額: {amount:>7}円 | 判定: {rank}")

# 3. 特定条件の抽出（論理演算子の復習）
# 4. 「一般顧客」かつ「15000円以下」の、キャンペーン対象者を探す
print("\n--- キャンペーン対象者（10,000～15,000円) ---")
for customer in customer_data:
    if customer[1] >= 10000 and customer[1] <= 15000:
        print(f"DM送付対象: {customer[0]}")
        
```
