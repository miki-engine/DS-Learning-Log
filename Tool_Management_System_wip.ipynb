# 道具管理システム

```python
from datetime import datetime, date
td = date.today()

with open("day10_data.csv", encoding="utf-8") as f:
    f.readline()
    for row in f:
        row = row.rstrip()
        cols = row.split(",")
        item = cols[1]
        last_used_date = datetime.strptime(cols[4], "%Y-%m-%d").date()

        # 未使用期間
        days_unused = (td - last_used_date).days

        if cols[5] == "broken":
            print(f"【故障中】廃棄・更新検討リスト: {item}")
        elif days_unused >= 180:
            print(f"【長期未使用機材】レンタル検討リスト: {item}:{days_unused}日未使用")
```
