# 道具管理システム

```python
from datetime import datetime, date
import unicodedata

def get_width(text):
    count = 0
    for char in str(text):
        if unicodedata.east_asian_width(char) in 'FWA':
            count += 2
        else:
            count += 1
    return count

td = date.today()

print(f"--- 本日の日付:{td}")
print("-" * 50)

target_width = 30
h_item = "道具名"
h_padding = " " * (target_width - get_width(h_item))
print(f"{h_item}{h_padding} | 未使用期間 | ステータス")
with open("tool_data.csv", encoding="utf-8") as f:
    f.readline()
    for row in f:
        row = row.rstrip()
        cols = row.split(",")
        item_name = cols[1]
        item_padding = " " * (target_width - get_width(item_name))
        last_used_date = datetime.strptime(cols[4], "%Y-%m-%d").date()

        # 未使用期間
        days_unused = (td - last_used_date).days
        days_str = f"{days_unused:>5}日"
        days_padding = " " * (10 - get_width(days_str))

        if cols[5] == "broken":
            status = "故障中"
        elif days_unused >= 180:
            status = "長期未使用道具"
        else:
            status = "通常使用"

        print(f"{item_name}{item_padding} | {days_str}{days_padding} | {status}")

print("-" * 50)
print("【故障中】廃棄・更新検討")
print("【長期未使用機材】レンタル検討")
```
