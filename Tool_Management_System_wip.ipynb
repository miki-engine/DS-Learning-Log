# 道具管理システム

```python
from datetime import datetime, date
td = date.today()

print(f"--- 本日の日付:{td}")
print("-" * 50)
print(f"{'道具名':<18} | {'未使用期間'} | {'ステータス'}")
with open("tool_data.csv", encoding="utf-8") as f:
    f.readline()
    for row in f:
        row = row.rstrip()
        cols = row.split(",")
        item = cols[1]
        last_used_date = datetime.strptime(cols[4], "%Y-%m-%d").date()

        # 未使用期間
        days_unused = (td - last_used_date).days

        if cols[5] == "broken":
            status = "故障中"
        elif days_unused >= 180:
            status = "長期未使用道具"
        else:
            status = "通常使用"

        print(f"{item:<18} | {days_unused}日 | {status}")

print("-" * 50)
print("【故障中】廃棄・更新検討")
print("【長期未使用機材】レンタル検討")
```
