# 演算子と条件分岐によるデータ抽出

```python
# 1. データの準備(商品名と売上のリスト)
sales_data = [
    ["Plushie", 5000],
    ["Game Tile1", 7800],
    ["Gaming Monitor", 45000],
    ["Keycap", 1500],
    ["Controller", 12000]
]

# 2. 判定基準 （比較演算子)
threshold = 10000 # 1万円以上の高額商品を抜き出したい

print(f"--- {threshold}円以上の商品リスト ---")

# 3. ループとif文(演算子)の組み合わせ
for item in sales_data:
    name = item[0]
    price = item[1]

    # 比較演算子
    if price >= threshold:
        print(f"高額商品: {name} ({price}円)")
    else:
        print(f"通常商品: {name}")

# 4. 論理演算子(and)の練習
# 5. 5000円以上 かつ 20000円以下の「手頃なアイテム」を探す
print("\n-- 予算内 (5000円～20000円)のアイテム ---")
for item in sales_data:
    if item[1] >= 5000 and item[1] <=20000:
        print(f"おすすめ:{item[0]}")
        
```
