# 時間帯別の売上集計

```python
# 購入ログデータ (Purchase Logs)
# time: 購入時刻(HH:MM), amount: 購入金額
purchase_logs = [
    {"time": "09:15", "amount": 1200},
    {"time": "12:30", "amount": 3500},
    {"time": "18:45", "amount": 5000},
    {"time": "12:15", "amount": 2000},
    {"time": "09:45", "amount": 1500},
    {"time": "19:20", "amount": 8000},
    {"time": "21:10", "amount": 4000},
    {"time": "12:50", "amount": 1800},
    {"time": "18:10", "amount": 3000},
]

hour_amount = {}
for log in purchase_logs:
    hour = log["time"][:2]
    if hour in hour_amount:
        hour_amount[hour] += log["amount"]
    else:
        hour_amount[hour] = log["amount"]

print(hour_amount)

```
