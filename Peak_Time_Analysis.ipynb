# 時間帯別の売上集計

```python
# 購入ログデータ (Purchase Logs)
# time: 購入時刻(HH:MM), amount: 購入金額
purchase_logs = [
    {"time": "09:15", "amount": 1200},
    {"time": "12:30", "amount": 3500},
    {"time": "18:45", "amount": 5000},
    {"time": "12:15", "amount": 2000},
    {"time": "09:45", "amount": 1500},
    {"time": "19:20", "amount": 8000},
    {"time": "21:10", "amount": 4000},
    {"time": "12:50", "amount": 1800},
    {"time": "18:10", "amount": 3000},
]

hour_amount = {} # {時間帯(HH): 売上合計(円)}
for log in purchase_logs:
    hour = log["time"][:2]
    if hour in hour_amount:
        hour_amount[hour] += log["amount"]
    else:
        hour_amount[hour] = log["amount"]

print("--- 時間帯別の売上合計 ---")

for h, amt in sorted(hour_amount.items()):
    print(f"{h}:00 - {h}:59 | 売上合計: {amt:5,d} 円")

best_hour = max(hour_amount, key=lambda x: hour_amount[x])

print("=" * 50)
print(f"ピーク時間帯: {best_hour}時台 ({hour_amount[best_hour]}円)")
print("=" * 50)
print("Analysis Insight:")
print(f"The highest sales were recorded during the {best_hour}:00 slot.")
print("We should consider increasing staff during this peak period.")
```
