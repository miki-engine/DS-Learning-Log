# 商品売上のABCランク分類

```python
sales_data = [
    {"item": "Monitor A", "price": 35000, "quantity": 10},
    {"item": "Mouse B", "price": 5000, "quantity": 50},
    {"item": "Keyboard C", "price": 12000, "quantity": 20},
    {"item": "Laptop D", "price": 150000, "quantity": 5},
    {"item": "Cable E", "price": 1500, "quantity": 100},
    {"item": "Webcam F", "price": 8000, "quantity": 15},
    {"item": "Desk G", "price": 45000, "quantity": 3},
    {"item": "Earphone H", "price": 25000, "quantity": 12},
]

# 1. 各商品の売上を計算して追加
for item in sales_data:
    item["sales"] = item["price"] * item["quantity"]

# 2. 総売上額を計算
total_sales = sum(item["sales"] for item in sales_data)

# 3.売上で降順に並び替え
sorted_data = sorted(sales_data, key=lambda x: x["sales"], reverse=True)

current_sum = 0
for item in sorted_data:
    current_sum += item["sales"]
    ratio = current_sum / total_sales
    if ratio <= 0.7:
        item["rank"] = "A"
    elif ratio <= 0.9:
        item["rank"] = "B"
    else:
        item["rank"] = "C"

print("=" * 50)
print("商品売上 ABC分析結果")
print("=" * 50)

ranks = {"A": "主力商品", "B": "準主力商品", "C": "その他"}
for rank_key, rank_label in ranks.items():
    print(f"\n【{rank_label}(ランク{rank_key})】")
    for item in sorted_data:
        if item["rank"] == rank_key:
            share = item["sales"] /total_sales * 100
            print(f" {item['item']:<12} 売上: {item['sales']:>8,}円 ({share:.1f}%)")

print(f"\n")
print("=" * 50)
print("総売上額")
print("=" * 50)
print(f"{total_sales:,}円")
            


```
